```python
import mysql.connector
import matplotlib.pyplot as plt

# Connect to MySQL
cnx_rapAPI = mysql.connector.connect(
    host='xx',
    user='xx',
    password='xxx',
    database='rapAPI2'
)

# Create a cursor
cursor = cnx_rapAPI.cursor()

# SQL query
query = """
SELECT 
    CASE
        WHEN service_level >= 10 AND service_level < 91 THEN '10-90'
        WHEN service_level >= 91 AND service_level < 96 THEN '91-95'
        WHEN service_level >= 96 AND service_level < 97 THEN '96'
        WHEN service_level >= 97 AND service_level < 98 THEN '97'
        WHEN service_level >= 98 AND service_level < 99 THEN '98'
        WHEN service_level >= 99 AND service_level < 100 THEN '99'
        WHEN service_level = 100 THEN '100'
    END AS service_level_range,
    AVG(popularity) AS avg_popularity
FROM 
    api_info
    where popularity > 0
GROUP BY 
    service_level_range
"""

# Execute the query
cursor.execute(query)

# Fetch all the results
results = cursor.fetchall()

# Close cursor and connection
cursor.close()
cnx_rapAPI.close()

# Process the fetched data
service_levels = []
avg_popularity = []

# Define the sequence of service level ranges
sequence = ['10-90', '91-95', '96', '97', '98', '99', '100']

# Populate data in the sequence
for item in sequence:
    for row in results:
        if row[0] == item:
            service_levels.append(row[0])
            avg_popularity.append(row[1])
            break

# Plot the data as a line graph
plt.figure(figsize=(4, 4))
plt.plot(service_levels, avg_popularity, color='blue', linestyle='-')
plt.xlabel('Service Level Range (%)')
plt.ylabel('Average Popularity')
plt.title('Avg Popularity by Service Level Range')
plt.xticks(fontsize=7)
plt.grid(True, linestyle='--', linewidth=0.5, color='lightgray')
plt.tight_layout()
plt.show()
```

```python
import mysql.connector
import matplotlib.pyplot as plt

# Connect to MySQL
cnx_rapAPI = mysql.connector.connect(
    host='xx',
    user='xx',
    password='xx',
    database='rapAPI2'
)

# Create a cursor
cursor = cnx_rapAPI.cursor()

# SQL query for popularity
popularity_query = """
SELECT 
    CASE
        WHEN service_level >= 10 AND service_level < 91 THEN '10-90'
        WHEN service_level >= 91 AND service_level < 96 THEN '91-95'
        WHEN service_level >= 96 AND service_level < 97 THEN '96'
        WHEN service_level >= 97 AND service_level < 98 THEN '97'
        WHEN service_level >= 98 AND service_level < 99 THEN '98'
        WHEN service_level >= 99 AND service_level < 100 THEN '99'
        WHEN service_level = 100 THEN '100'
    END AS service_level_range,
    AVG(popularity) AS avg_popularity
FROM 
    api_info
WHERE 
    popularity > 0
GROUP BY 
    service_level_range
"""

# Execute the popularity query
cursor.execute(popularity_query)
popularity_results = cursor.fetchall()

# SQL query for repository count
repo_count_query = """
SELECT
    service_level_range,
    AVG(repo_count) AS avg_repo_count
FROM
    (SELECT
        CASE
            WHEN ai.service_level >= 10 AND ai.service_level < 91 THEN '10-90'
            WHEN ai.service_level >= 91 AND ai.service_level < 96 THEN '91-95'
            WHEN ai.service_level >= 96 AND ai.service_level < 97 THEN '96'
            WHEN ai.service_level >= 97 AND ai.service_level < 98 THEN '97'
            WHEN ai.service_level >= 98 AND ai.service_level < 99 THEN '98'
            WHEN ai.service_level >= 99 AND ai.service_level < 100 THEN '99'
            WHEN ai.service_level = 100 THEN '100'
        END AS service_level_range,
        gd.hostname,
        COUNT(gd.repository_name) AS repo_count
    FROM
        api_info ai
    INNER JOIN
        github_details gd ON ai.hostname = gd.hostname
    GROUP BY
        service_level_range,
        gd.hostname) AS subquery
GROUP BY
    service_level_range;
"""

# Execute the repository count query
cursor.execute(repo_count_query)
repo_count_results = cursor.fetchall()

# Close cursor and connection
cursor.close()
cnx_rapAPI.close()

# Process fetched data for popularity
popularity_dict = {row[0]: row[1] for row in popularity_results}

# Process fetched data for repository count
repo_count_dict = {row[0]: row[1] for row in repo_count_results}

# Define the sequence of service level ranges
sequence = ['10-90', '91-95', '96', '97', '98', '99', '100']

# Initialize lists to store data
service_levels = []
avg_popularity = []
avg_repo_counts = []

# Populate data in the sequence
for item in sequence:
    service_levels.append(item)
    avg_popularity.append(popularity_dict.get(item, 0))
    avg_repo_counts.append(repo_count_dict.get(item, 0))

# Plotting
fig, ax1 = plt.subplots(figsize=(5, 4))

# Plot for popularity
color = 'blue'
ax1.set_xlabel('Service Level Range')
ax1.set_ylabel('Average Popularity', color=color)
ax1.plot(service_levels, avg_popularity, color=color, label='RapidAPI popularity')
ax1.tick_params(axis='y', labelcolor=color)
ax1.tick_params(axis='x', labelsize=8)

# Create a secondary y-axis
ax2 = ax1.twinx()

# Plot for repository count
color = 'darkorange'
ax2.set_ylabel('Average Repository Count', color=color)
ax2.plot(service_levels, avg_repo_counts, color=color, label='GitHub popularity')
ax2.tick_params(axis='y', labelcolor=color)

# Finalize plot
plt.title('Avg Popularity, Repository Count by Service Level Range', fontsize = 10)
ax1.legend(loc='lower center', bbox_to_anchor=(0.52, 0.1))  # Legend for RapidAPI popularity
ax2.legend(loc='lower center')  # 
plt.grid(True, linestyle='--', linewidth=0.5, color='lightgray')

plt.xticks(fontsize=7)
plt.tight_layout()
plt.savefig(r'D:\RapidAPI\RapidAPI\ServiceLevelVsPop.pdf', format='pdf')

plt.show()
```
