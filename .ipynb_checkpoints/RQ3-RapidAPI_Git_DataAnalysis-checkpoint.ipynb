```python
import pandas as pd

# Read the Excel file and specifically 'Sheet1'
df = pd.read_excel('RapidAPI_1.xlsx', sheet_name='Sheet1')

# Adjust the DataFrame to consider the first row as header by using header=0 (the default)
# Then we can directly use the resulting DataFrame without skipping rows manually
df = pd.read_excel('RapidAPI_1.xlsx', sheet_name='Sheet1', header=0)

# Initialize an empty dictionary to store your data
services_dict = {}

# Iterate over the DataFrame from row 2 to row 380
# Note: iloc in pandas is 0-indexed, so row 2 is at index 1, row 380 is at index 379
for index in range(0, 380):  # This will go from the second row (index 1) to the 380th row (index 379)
    # Check if the row is within the DataFrame (in case there are fewer than 380 rows)
    if index < len(df):
        row = df.iloc[index]
        key = row.iloc[0]  # Get the first cell in the row as the key
        values = row.iloc[1:].tolist()  # Get the rest of the row as values
        services_dict[key] = values  # Assign to dictionary

# Now, services_dict contains all your keys and values from row 2 to row 380
print(services_dict)
```

```python
import mysql.connector


# Connect to MySQL
cnx_rapAPI = mysql.connector.connect(
    host='xxx',
    user='xx',
    password='xxx',
    database='rapAPI2'
)
cursor = cnx_rapAPI.cursor()

# SQL INSERT statement
add_service = ("INSERT INTO equi_services "
               "(hostname, service1, service2, service3, service4, service5) "
               "VALUES (%s, %s, %s, %s, %s, %s)")

# Iterate over the dictionary and execute the INSERT statement for each item
for selected_service, services in services_dict.items():
    services = [(None if pd.isnull(service) else service) for service in services]

    # Ensure there are exactly 5 service URLs (or however many your table supports)
    services += [None] * (5 - len(services))  # Pad the list if fewer than 5 services
    data_service = (selected_service, *services[:5])  # Get only the first 5 services
    cursor.execute(add_service, data_service)

# Make sure data is committed to the database
cnx_rapAPI.commit()

# Close the cursor and connection
cursor.close()
cnx_rapAPI.close()
```

```python
import mysql.connector

# Connect to MySQL
cnx = mysql.connector.connect(
    host='141.215.80.233',
    user='root',
    password='SiddhiB@123',
    database='rapAPI2'
)

# Create a cursor
cursor = cnx.cursor()

# Fetch selected base URLs
cursor.execute("SELECT selected_base_url FROM equi_services")
selected_services = [row[0] for row in cursor.fetchall()]


cursor.execute("SELECT service1 FROM equi_services where service1 is not null")
service1 = [row[0] for row in cursor.fetchall()]

cursor.execute("SELECT service2 FROM equi_services where service2 is not null")
service2 = [row[0] for row in cursor.fetchall()]

cursor.execute("SELECT service3 FROM equi_services where service3 is not null")
service3 = [row[0] for row in cursor.fetchall()]

cursor.execute("SELECT service4 FROM equi_services where service4 is not null")
service4 = [row[0] for row in cursor.fetchall()]

cursor.execute("SELECT service5 FROM equi_services where service5 is not null")
service5 = [row[0] for row in cursor.fetchall()]

optional_services_set = set()

# Iterating through each service column and fetching unique values
for i in range(1, 6):  # From service1 to service5
    query = f"SELECT service{i} FROM equi_services WHERE service{i} IS NOT NULL"
    cursor.execute(query)
    services = [row[0] for row in cursor.fetchall()]
    optional_services_set.update(services)

# Convert the set to a list
optional_services = list(optional_services_set)
print(len(optional_services))

# Initialize lists for selected and optional data
selected_popularity = []
selected_latency = []
selected_service_level = []
selected_pricing = []
selected_doc = []
selected_com = []

optional_popularity = []
optional_latency = []
optional_service_level = []
optional_pricing = []
optional_doc = []
optional_com = []

# Fetch data for selected base URLs
for base_url in selected_services:
    # Fetch popularity
    cursor.execute("SELECT popularity FROM api_info WHERE base_url = %s", (base_url,))
    result = cursor.fetchone()
    if result is not None:
        selected_popularity.append(result[0])
    
    # Fetch latency
    cursor.execute("SELECT latency FROM api_info WHERE base_url = %s", (base_url,))
    result = cursor.fetchone()
    if result is not None:
        selected_latency.append(result[0])
    
    # Fetch service level
    cursor.execute("SELECT service_level FROM api_info WHERE base_url = %s", (base_url,))
    result = cursor.fetchone()
    if result is not None:
        selected_service_level.append(result[0])
    
    # Fetch pricing
    cursor.execute("SELECT calculated_cost FROM api_pricing WHERE base_url = %s", (base_url,))
    result = cursor.fetchone()
    if result is not None:
        selected_pricing.append(result[0])
    
    # Fetch doc support
    cursor.execute("SELECT service_provider_support FROM api_details WHERE base_url = %s", (base_url,))
    result = cursor.fetchone()
    if result is not None:
        selected_doc.append(result[0])
    
    # Fetch com support
    cursor.execute("SELECT support_weight FROM api_details WHERE base_url = %s", (base_url,))
    result = cursor.fetchone()
    if result is not None:
        selected_com.append(result[0])

# Fetch data for optional services
for service in optional_services:
    # Fetch popularity
    cursor.execute("SELECT popularity FROM api_info WHERE base_url = %s", (service,))
    result = cursor.fetchone()
    if result is not None:
        optional_popularity.append(result[0])
    
    # Fetch latency
    cursor.execute("SELECT latency FROM api_info WHERE base_url = %s", (service,))
    result = cursor.fetchone()
    if result is not None:
        optional_latency.append(result[0])
    
    # Fetch service level
    cursor.execute("SELECT service_level FROM api_info WHERE base_url = %s", (service,))
    result = cursor.fetchone()
    if result is not None:
        optional_service_level.append(result[0])
    
    # Fetch pricing
    cursor.execute("SELECT calculated_cost FROM api_pricing WHERE base_url = %s", (service,))
    result = cursor.fetchone()
    if result is not None:
        optional_pricing.append(result[0])
    
    # Fetch doc support
    cursor.execute("SELECT service_provider_support FROM api_details WHERE base_url = %s", (service,))
    result = cursor.fetchone()
    if result is not None:
        optional_doc.append(result[0])
    
    # Fetch com support
    cursor.execute("SELECT support_weight FROM api_details WHERE base_url = %s", (service,))
    result = cursor.fetchone()
    if result is not None:
        optional_com.append(result[0])

# Close cursor and connection
cursor.close()
cnx.close()

# Print the lengths of selected and optional data
print("Selected Services:")
print("Popularity:", len(selected_popularity))
print("Latency:", len(selected_latency))
print("Service Level:", len(selected_service_level))
print("Pricing:", len(selected_pricing))
print("Doc Support:", len(selected_doc))
print("Com Support:", len(selected_com))

print("\nOptional Services:")
print("Popularity:", len(optional_popularity))
print("Latency:", len(optional_latency))
print("Service Level:", len(optional_service_level))
print("Pricing:", len(optional_pricing))
print("Doc Support:", len(optional_doc))
print("Com Support:", len(optional_com))
```

```python
import matplotlib.pyplot as plt

# Define the popularity ranges
popularity_ranges = ['1-1.9', '2-2.9', '3-3.9', '4-4.9', '5-5.9', '6-6.9', '7-7.9', '8-8.9', '9-10']

# Function to count occurrences in a list for each range
def count_occurrences(popularity_list):
    counts = [0] * len(popularity_ranges)
    for pop in popularity_list:
        index = int(pop) if pop == 10 else int(pop // 1)
        if 1 <= index <= 9:
            counts[index - 1] += 1
    return counts

# Count occurrences in both lists
selected_counts = count_occurrences(selected_popularity)
optional_counts = count_occurrences(optional_popularity)

# Total count of each list
total_selected = sum(selected_counts)
total_optional = sum(optional_counts)

# Calculate percentages
selected_percentages = [count / total_selected * 100 if total_selected != 0 else 0 for count in selected_counts]
optional_percentages = [count / total_optional * 100 if total_optional != 0 else 0 for count in optional_counts]

# Plotting
x = range(len(popularity_ranges))
plt.figure(figsize=(4, 4))
plt.plot(x, selected_percentages, label='Selected services', color = 'blue')
plt.plot(x, optional_percentages, label='Optional services', color = 'darkorange')
plt.xlabel('Popularity range')
plt.ylabel('Percentage of services (%)')
plt.title('Services in popularity range')
plt.xticks(x, popularity_ranges, fontsize = 7)
plt.legend()
plt.ylim(0, 100) 
plt.yticks(range(0, 101, 25))
plt.grid(True, linestyle='--', linewidth=0.5, color='lightgray')
plt.tight_layout(pad=1.1)
plt.savefig(r'D:\RapidAPI\RapidAPI\SimilarServices_Popularity.pdf', format='pdf', dpi=600)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the latency ranges
latency_ranges = ['1-99', '100-299', '300-599', '600-999', '1000-9999', '10000-99999']

# Function to count occurrences in a list for each range
def count_occurrences(latency_list):
    counts = [0] * len(latency_ranges)
    for latency in latency_list:
        latency = int(latency)
        if latency >= 1 and latency <= 99:
            counts[0] += 1
        elif latency >= 100 and latency <= 299:
            counts[1] += 1
        elif latency >= 300 and latency <= 599:
            counts[2] += 1
        elif latency >= 600 and latency <= 999:
            counts[3] += 1
        elif latency >= 1000 and latency <= 9999:
            counts[4] += 1
        elif latency >= 10000 and latency <= 99999:
            counts[5] += 1
    return counts

# Count occurrences in both lists
selected_counts = count_occurrences(selected_latency)
optional_counts = count_occurrences(optional_latency)

# Total count of each list
total_selected = sum(selected_counts)
total_optional = sum(optional_counts)

# Calculate percentages
selected_percentages = [count / total_selected * 100 if total_selected != 0 else 0 for count in selected_counts]
optional_percentages = [count / total_optional * 100 if total_optional != 0 else 0 for count in optional_counts]

latency_ranges[-1] = '>10000'
# Plotting
x = range(len(latency_ranges))
plt.figure(figsize=(4, 4))
plt.plot(x, selected_percentages, label='Selected services', color = 'blue')
plt.plot(x, optional_percentages, label='Optional services', color = 'darkorange')
plt.xlabel('Latency range (ms)')
plt.ylabel('Percentage of services (%)')
plt.title('Services in latency range')
plt.xticks(x, latency_ranges, fontsize=7)
plt.legend()
plt.ylim(0, 100) 
plt.yticks(range(0, 101, 25))
plt.grid(True, linestyle='--', linewidth=0.5, color='lightgray')
plt.tight_layout()
plt.savefig(r'C:\Users\siddh\Desktop\RapidAPI\SimilarServices_Latency.pdf', format='pdf')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the pricing ranges
pricing_ranges = ['0', '0-0.01', '0.01-0.1', '0.1-1', '1-1000']

# Function to count occurrences in a list for each range
def count_occurrences(pricing_list):
    counts = [0] * len(pricing_ranges)
    for pricing in pricing_list:
        pricing = float(pricing)
        if pricing == 0:
            counts[0] += 1
        elif pricing > 0 and pricing <= 0.01:
            counts[1] += 1
        elif pricing > 0.01 and pricing <= 0.1:
            counts[2] += 1
        elif pricing > 0.1 and pricing <= 1:
            counts[3] += 1
        elif pricing > 1 and pricing <= 1000:
            counts[4] += 1
    return counts

# Count occurrences in both lists
selected_counts = count_occurrences(selected_pricing)
optional_counts = count_occurrences(optional_pricing)

# Total count of each list
total_selected = sum(selected_counts)
total_optional = sum(optional_counts)

# Calculate percentages
selected_percentages = [count / total_selected * 100 if total_selected != 0 else 0 for count in selected_counts]
optional_percentages = [count / total_optional * 100 if total_optional != 0 else 0 for count in optional_counts]

# Plotting
x = range(len(pricing_ranges))
plt.figure(figsize=(4, 4))
plt.plot(x, selected_percentages, label='Selected services', color = 'blue')
plt.plot(x, optional_percentages, label='Optional services', color = 'darkorange')
plt.xlabel('Pricing range (cost per request($))')
plt.ylabel('Percentage of services (%)')
plt.title('Services in pricing range')
plt.xticks(x, pricing_ranges, fontsize=8)
plt.legend()
plt.ylim(0, 100) 
plt.yticks(range(0, 101, 20))
plt.grid(True, linestyle='--', linewidth=0.5, color='lightgray')
plt.tight_layout()
plt.savefig(r'C:\Users\siddh\Desktop\RapidAPI\SimilarServices_Pricing.pdf', format='pdf')

plt.show()
```
